```python
all_domains = speaker_df.email.map(lambda x: [e.split('@')[1] for e in x.split(';')] if x else [])
```

```python
all_domains.head()
```

```python
import itertools
from collections import Counter
```

```python
flat_emails = itertools.chain.from_iterable(all_domains.values)
```

```python
Counter(flat_emails).most_common(20)
```

```python
all_tlds = speaker_df.website.map(lambda x: [e.split('.')[-1].split('/')[0] for e in x.split(';')] if isinstance(x, str) else [])
```

```python
flat_tlds = itertools.chain.from_iterable(all_tlds.values)
```

```python
Counter(flat_tlds).most_common(20)
```
